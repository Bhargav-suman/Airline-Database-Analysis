```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df['month'] = pd.Categorical(df['month'],categories=['January',"February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
], ordered=True)

df = df.sort_values(['year','month'])

plt.figure(figsize=(12, 6))
for year, group in df.groupby('year'):
    plt.plot(group['month'], group['passengers'], label=year)

plt.title("Monthly Passenger Traffic Trend for Each Year")
plt.xlabel("Month")
plt.ylabel("Number of Passengers")
#plt.xticks(rotation=45)
plt.legend(title="Year", bbox_to_anchor=(1.05, 1), loc='upper left')
#plt.tight_layout()
plt.show()
```

```python
monthly_data = df.groupby('month')['passengers'].mean()
#print(monthly_data)

plt.figure(figsize=(10, 5))
plt.plot(monthly_data.index, monthly_data.values, marker='o', linestyle='-', color='b')
plt.title("Overall Monthly Passenger Traffic Trend (All Years Combined)")
plt.xlabel("Month")
plt.ylabel("Average Number of Passengers")
#plt.xticks(rotation=45)
plt.grid(True)
#plt.tight_layout()
plt.show()
```
